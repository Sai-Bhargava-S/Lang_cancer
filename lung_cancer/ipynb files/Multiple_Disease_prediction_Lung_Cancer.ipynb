From the visualizations, it is clear that in the given dataset, the features GENDER, AGE, SMOKING and SHORTNESS OF BREATH don't have that much relationship with LUNG CANCER. So let's drop those features to make this dataset more clean.

```python
df_new=df.drop(columns=['GENDER','AGE', 'SMOKING', 'SHORTNESS OF BREATH'])
df_new
```

```python
#Splitting independent and dependent variables
X=df_new.iloc[:,:-1].values
y=df_new.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

## **Logestic regression**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
X_train_pred=lr.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,y_train)
print("accuracy on the training data",training_data_accuracy)
```

```python
#accuracy score on the test data
X_test_pred=lr.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,y_test)
print("accuracy on the test data",test_data_accuracy)
```
